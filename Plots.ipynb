# Plots

```python
import numpy as np
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 12)
import matplotlib.pyplot as plt
```

### Rectangular Patch

```python
plt.scatter((0.5), (0.5), s=100, c='green', label='Robot start')
plt.arrow((0.5), (0.5), (0.1), (0), capstyle='projecting', head_width=0.05)
plt.scatter((0.5), (2.5), s=100, c='cyan', label='Final goal')

plt.plot((3, 3), (0, 3), 'k--', label='Forward boundary')

plt.plot((0.5, 2.5), (0.75, 0.75), 'k-', label='Patch to clean')
plt.scatter((0.5), (0.75), s=100, c='red', label='Lower left')
plt.plot((0.5, 2.5), (2.5, 2.5), 'k-')
plt.plot((0.5, 0.5), (0.75, 2.5), 'k-',)
plt.plot((2.5, 2.5), (0.75, 2.5), 'k-')

plt.xticks(np.arange(0, 3.5, 0.25))
plt.yticks(np.arange(0, 3.5, 0.25))
plt.legend()
plt.show()
```
